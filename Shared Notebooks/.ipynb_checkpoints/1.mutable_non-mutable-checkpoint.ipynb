## Mutable (value of a object can be changed after it has been created)
- List
- Dictionary
- Set

```python
# List
lsts = [1, 2, 3]
print(id(lsts))  # Memory address before change

lsts.append(4)   # Modify the list
print(lsts)
print(id(lsts)) # Memory address after change
```

```python
person = {'name': 'Toto', 'age': 20}
print(id(person))  # Memory address before change

person['age'] = 26
print(person)
print(id(person))  # Memory address after change
```

## Immutable (value of a object cannot be changed after it has been created)
- int
- float
- bool
- str
- tuple

```python
# Integer
x = 0
print(id(x))  # memory address of x

x = 5
print(id(x))
```

```python
# Float
y = 0.0
print(id(y))  # memory address of y

y += 5.1
print(id(y))
```

```python
# Tuple
t = (1, 2, 3)
print(id(t))

t[0] = 100
```
